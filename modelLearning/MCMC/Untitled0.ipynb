```python
%matplotlib inline
import pymc.Matplot as pymcPlt
import pymc
import matplotlib.pyplot as plt
import matplotlib
import numpy as np
matplotlib.rcParams['figure.figsize'] = (10,10)
np.set_printoptions(formatter={'float': '{: 0.3f}'.format})
```

```python
theta = pymc.Dirichlet('Dir', theta=[10, 10, 10])
theta.value
```

```python
Y = pymc.Multinomial('Multi', n = 1000, p = theta, value = [500, 500, 0], observed=True)
```

```python
M = pymc.MCMC([theta, Y])
```

```python
M.sample(iter=10000)
```

```python
theta.value
```

```python
pymcPlt.plot(M)
```

```python
gamma = pymc.Dirichlet('Dir2', theta=[10, 10, 10])
gamma.value
```

```python
Z = pymc.Multinomial('Multi2', n = 1000, p= gamma) #= [.2, .4, .4]) #p=gamma
```

```python
@pymc.potential
def pot(Z=Z):
    return -(Z[0]-500)**2 - (Z[1]-500)**2 - (Z[2]-0)**2
    
```

```python
M2 = pymc.MCMC([gamma, Z, pot])
```

```python
M2.sample(iter=50000, thin=10)
```

```python
gamma.value
```

```python
pymcPlt.plot(M2)
```
