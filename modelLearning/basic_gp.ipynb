
We have data with noise, except that in a certain region, the data comes from two different sources.

The goal is to recover the **region** of data where there is this high variance (here caused by different sources, but could be because of high sensor noise)

```python
N = 24
X1 = np.random.uniform(-5.,5.,N)
err1 = np.random.randn(N)*0.05;
Y1 = np.sin(X1) + err1;

X2 = np.random.uniform(-5.,5.,N)
Y2 = np.sin(X2)
#Between -4, -1, there is some data from cosine
idx = np.logical_and(X2 < -1, X2 > -4)
Y2[idx] = np.cos(X2[idx])

err2 = np.random.randn(N)*0.05;
Y2 = Y2 + err2

X = np.concatenate((X1,X2))
Y = np.concatenate((Y1,Y2))

error = np.concatenate((err1,err2))
```

```python
kernel = GPy.kern.RBF(input_dim=1, variance=1., lengthscale=1.)
```

```python
m = GPy.models.GPRegression(X[:,None],Y[:,None],kernel)
m.optimize_restarts(num_restarts = 10);
```

```python
display(m)
_ = m.plot()
```
