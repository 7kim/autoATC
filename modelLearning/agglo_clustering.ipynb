```julia
trainset_idx = randsubseq(1:size(XBig,1),5000./size(XBig,1));
Xtrain = XBig[trainset_idx, :];
```

```julia
# Create a graph capturing local connectivity. Larger number of neighbors
# will give more homogeneous clusters to the cost of computation
# time. A very large number of neighbors gives more evenly distributed
# cluster sizes, but may not impose the local manifold structure of
# the data
knn_graph = skN.kneighbors_graph(Xtrain[:,1:2], 30, include_self=false)
```

```julia
aggCluster = skC.AgglomerativeClustering(n_clusters=20, affinity="manhattan", linkage="average") #linkage="ward", 
ward = aggCluster[:fit](Xtrain)
```

```julia
label = ward[:labels_];
label_unique = sort(unique(label));
```

```julia
markers = itertools.cycle(["x","+",".","o"])

for (l, m) in zip(label_unique,  markers)
    #Get color
    c = PyPlot.plt.cm[:spectral][:__call__](np.float(l) / np.max(label_unique + 1))
    
    #Get data, swap e/n for plotting
    Xi = Xtrain[find(x-> x== l, label),1:2]';

    #Plot scatter of data
    PyPlot.plot(Xi[1,:], Xi[2,:], linestyle="", markersize=4, marker=m, alpha=0.8, color=c)

        
end

grid("on")
axis("equal")
```
