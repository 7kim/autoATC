```julia
function PiSimple(ss, perm = [])
    a = noaction
    if perm == []
        perm = [1:length(ss)]
    end
    ss_perm = sortperm([sn[s] for s in ss])
    ss_sorted = ss[ss_perm]
    if(ss_sorted != ss)
        return PiSimple(ss_sorted, ss_perm)
    else        
        for AC1 in ss
            amax = noaction
            rmax = Reward(ss,amax)
            for av in validActions(ss)[2:end] #1st action is always no action
                for ssnext in getNextPerms(ss)
                    if(ssnext[av[1]] == av[2]) #Limit to states that have this action
                        r = Reward(ss,amax) + Reward(ssnext,noaction)
                        if r > rmax
                           rmax = r
                           amax = av
                        end
                    end
                end
            end
            if amax != noaction
                a = (perm[amax[1]] , amax[2])
                break#We picked an action, stop
            end            
        end
        #If we have not picked an action still
        #Make sure we push aircraft out of the taxi state
        if a == noaction
            for AC1 in ss
                 if AC1 == :T
                    noconflict = true
                    for s2 in ss[2:end]
                        noconflict = (s2 == :T) || !(:R in NextStates[s2])
                        if !noconflict 
                            break
                        end
                    end
                    if noconflict
                        a = (perm[1], :R)
                        break;
                    end
                end
            end
        end
    end
    
    return a
end
# PiS = Dict();
# for s in S[:]
#     PiS[s] = PiSimple(s)
# end
```
