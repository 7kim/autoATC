```julia
function PiSimple(ss, perm = [])
    a = noaction
    if perm == []
        perm = [1:length(ss)]
    end
    ss_perm = sortperm([sn[s] for s in ss])
    ss_sorted = ss[ss_perm]
    if(ss_sorted != ss)
        return PiSimple(ss_sorted, ss_perm)
    else
        AC1 = ss[1]
        AC1n = NextStates[AC1];
        if AC1 == :T
            noconflict = true
            for s2 in ss[2:end]
                noconflict = (s2 == :T) || !(:R in NextStates[s2])
                if !noconflict 
                    break
                end
            end
            if noconflict
                a = (perm[1], :R)
            end
        else
            amax = noaction
            rmax = Reward(ss,amax)
            for av in validActions(ss)
                if av != noaction
                    ssnext = deepcopy(ss)
                    for i in 1:length(ssnext)
                        ssnext[i] = NextStates[ssnext[i]][1]
                    end
                    ssnext[av[1]] = av[2]
                    r = Reward(ss,amax) + Reward(ssnext,noaction)
                    if r > rmax
                       rmax = r
                       amax = av
                    end
                end
            end
            if amax != noaction
                a = (perm[amax[1]] , amax[2])
            end            
        end
    end
    
    return a
end
PiS = Dict();
for s in S[:]
    PiS[s] = PiSimple(s)
end
```
