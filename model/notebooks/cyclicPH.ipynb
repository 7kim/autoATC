```julia
function erlang(μ1, c, N)
    exp1 = Exponential(μ1)
    r = zeros(N)
    for i in 1:N
        r[i] += sum(rand(exp1, c))        
    end
    return r
end
function cyclicCox(μ1, μ2, p1, p2, N)
    r = zeros(N)
    exp1 = Exponential(μ1);
    exp2 = Exponential(μ2);
    for i in 1:N
        r[i] = cyclicCox(exp1, exp2, p1, p2);
    end
    
    return r
end

function cyclicCox(d1, d2, p1, p2)
    assert( p2 < 1)
    t = 0.;
    #Sample from first node
    t += rand(d1)
    if rand() < p1
        #second state, sample
        t += rand(d2)
        if rand() < p2
            #cycle!
            t += cyclicCox(d1,d2,p1,p2)
        end
    end
    
    return t
    
end
```

```julia
x = Vector{Float64}[];
p2list = [0., .1, 0.25, 0.5, 0.75]; np = length(p2list)
mu2 = [.4, .8]; nm = length(mu2)

k = 0
for mu in mu2
    for i in 1:np
        k += 1
        if k <= np
            push!(x, cyclicCox(1, mu, 1, p2list[i], 5000));
        else
            push!(x, erlang(1, i, 5000));
        end
        x[k] = x[k]/mean(x[k])
    end
end
```

```julia
histData = {}
k = 0
for j in 1:nm
    for i in 1:np
        k += 1
        push!(histData, PyPlot.hist(x[k],20, histtype="step", normed=true, align="right"))
    end
end

PyPlot.clf()

k = 0
for j in 1:nm
    c = j == 1 ? "blue" : "red";
    for i in 1:np
        k += 1
        plot([histData[k][2][1:end-1]],histData[k][1],color=c,linestyle="--")
    end
end
grid("on")
```

```julia
m1 = 1.; m2 =1.; p1 = 0.8; p2 = 0.3;
y = cyclicCox(m1,m2, p1, p2, 500000);
mean(y), std(y), std(y)/mean(y)
```

```julia
z = cyclicCox(m1, (m2 + p2*m1)/(1-p2), p1, 0., 500000);
mean(z), std(z), std(z)/mean(z)
```
