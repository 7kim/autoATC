```julia
events = find(diff(xHist));
eventLength = [events[1], diff(events)]*sim3d.simdt;
eventStates = xHist[events];
```

```julia
averages = zeros(length(unique(eventStates)))
variances = zeros(length(unique(eventStates)))
cdfs = Dict{pattern.sType,pattern.cdfTime}()
```

```julia
astr = string(:LD1)
```

```julia
for x in 1:length(averages)
    s = pattern.x2s(x)
    astr = string(s)
    astr_ = copy(astr)
    if string(s) == "R"
        nothing    
    elseif(astr[1] == 'R')
        astr_ = replace(astr, 'R', 'L')
    elseif(astr[1] == 'L')
        astr_ = replace(astr, 'L', 'R')
    end
    
    s2 = symbol(astr_)
    x2 = pattern.s2x(s2)
    idx = find(e -> e == x || e == x2, eventStates)
    
    
    tEvents = eventLength[idx]
    averages[x] = mean(tEvents)
    variances[x] = var(tEvents)
    

    n = min(max(int(length(tEvents)/2), 5),50)
    (times, cnts) = hist(tEvents, linspace(minimum(tEvents),maximum(tEvents),n))
    probs = cumsum(cnts); probs /= probs[end]; insert!(probs, 1, 0.)

    cdfs[s] = pattern.cdfTime(probs, times)
    
end
cv2 = variances ./ (averages.^2);
```

```julia
fout = open("patternTimes.jl","w")

write(fout,"sojurnTime = (sType => Float32)[]\n")
for i in 1:length(averages)
    s = pattern.x2s(i)
    write(fout, "sojurnTime[:$s]=$(averages[i])\n") 
end


write(fout,"sojurnCDF = Dict{sType, cdfTime}()\n")
for (s, c) in cdfs
    write(fout,"sojurnCDF[:$s] = cdfTime($(c.probs), $(c.times))\n")
end

write(fout,"\n\n")

close(fout)
```

```julia
function randSum(D, N)
    x = zeros(N,1);    
    for d in D
        x = x+rand(d,N)  
    end
    return x
end

function randCrazy(avg, N)
    d1 = Exponential(avg/2)
    d2 = Exponential(avg*2)
    return [rand(d1,int64(2*N/3)), rand(d2,int64(N/3))]
end
    
    
```

```julia
function sampleTime(cdf::pattern.cdfTime ,tclip::Float32, p::Float32)
    N = length(cdf.probs)
    
    #adjust to clip time    
    if tclip > 0
        cnt = 0
        for t in cdf.times
            cnt += 1
            if t >= tclip
                break
            end
        end
        p = cdf.probs[cnt] + p * (1f0 - cdf.probs[cnt])
    end

    idx_L = searchsortedlast(cdf.probs, p)
    idx_R = min(idx_L+1, N)

    frac = max((p - cdf.probs[idx_L])/(cdf.probs[idx_R] - cdf.probs[idx_L]), 0.)
    frac = min(frac, 1.)

#    @show idx_L, cdf.probs[idx_L], idx_R, cdf.probs[idx_R],  frac
    t = cdf.times[idx_L] + frac * (cdf.times[idx_R] - cdf.times[idx_L]) 

    return max(t, tclip)
end
```
