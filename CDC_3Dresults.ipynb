```julia
data = {load("./results/3DSimResults_n"*string(n)*".jld") for n in 1:4}
nBetas = size(data[1]["flightTimes"],1)
N = size(data[1]["flightTimes"],3)
#"tStat", tStat, "alertRate", alertRates, "flightTimes", flightTimes);
```

```julia

ft_stats = zeros(nBetas+1, 2) #mean and std
alert_stats = similar(ft_stats)

for n in 1:length(data)
    for b in 1:(nBetas+1)
        atcIdx = 1
        betaIdx = b
        if b > nBetas
            atcIdx = 2 #Silent ATC
            betaIdx = 1
        end
        ft = data[n]["flightTimes"][betaIdx,atcIdx,:][:] / 3600
        alerts = data[n]["alertRate"][betaIdx,atcIdx,:][:]
        
        ft_stats[b,:]    = [mean_and_std(ft)...]
        alert_stats[b,:] = [mean_and_std(alerts)...]
    end
    #sort by alert means
    p = sortperm(alert_stats[:, 1]) #sorting using the mean
    alert_stats = alert_stats[p, :]
    ft_stats = ft_stats[p, :]
    
#     xerrL = alert_stats[:,2] - alert_stats[:,3]
#     xerrR = alert_stats[:,4] - alert_stats[:,3]
    
#     yerrL = ft_stats[:,2] - ft_stats[:,3]
#     yerrR = ft_stats[:,4] - ft_stats[:,3]
    xerrL = alert_stats[:,2]/sqrt(N)
    xerrR = alert_stats[:,2]/sqrt(N)
    yerrL = ft_stats[:,2]/sqrt(N)
    yerrR = ft_stats[:,2]/sqrt(N)
    
    
    PyPlot.errorbar(alert_stats[:, 1], ft_stats[:, 1], 
    xerr = [xerrL xerrR ]',  marker="o", linestyle = "--", yerr = [yerrL yerrR]')
end
PyPlot.ylim([0, 5])
PyPlot.grid("on")
PyPlot.xlabel("Alert Rate (commands/minute)")
PyPlot.ylabel("Flight Hours before NMAC (hours)")
PyPlot.legend(["c="*string(n) for n in 1:4], bbox_to_anchor=(0.25, 1.), loc = 9, ncol=2, numpoints=1)
plt.savefig("./results/Flight_hours_VS_alert_rates.pgf", bbox_inches="tight")
```

```julia
data = {load("./results/new_3DSimResults_n"*string(n)*".jld") for n in 1:5}
nBetas = size(data[1]["flightTimes"],1)
N = size(data[1]["flightTimes"],3)

# "tStat", tStat, "alertCounts", alertCounts, 
# "flightTimes", flightTimes, "tTotals", tTotals, "nNMACcounts", nNMACcounts);
```

```julia
nmac_stats = zeros(nBetas+1, 2) #mean and std
alert_stats = similar(nmac_stats)


minN = Inf
maxN = -Inf
for n in 1:length(data)
    for b in 1:(nBetas+1)
        atcIdx = 1
        betaIdx = b
        if b > nBetas
            atcIdx = 2 #Silent ATC
            betaIdx = 1
        end
        ft = data[n]["flightTimes"][betaIdx,atcIdx,:][:]/3600
        nmacsPerHour = data[n]["nNMACcounts"][betaIdx,atcIdx,:][:] ./ ft
        alertsPerHour = data[n]["alertCounts"][betaIdx,atcIdx,:][:] ./ ft
        
        nmac_stats[b,:]    = [mean_and_std(nmacsPerHour)...]
        alert_stats[b,:] = [mean_and_std(alertsPerHour/60)...]
    end
    #sort by alert means
    p = sortperm(alert_stats[:, 1]) #sorting using the mean
    alert_stats = alert_stats[p, :]
    nmac_stats = nmac_stats[p, :]
    
    minN = min(minN, minimum(nmac_stats[:,1]))
    maxN = max(maxN, maximum(nmac_stats[:,1]))
    
#     xerrL = alert_stats[:,2] - alert_stats[:,3]
#     xerrR = alert_stats[:,4] - alert_stats[:,3]
    
#     yerrL = ft_stats[:,2] - ft_stats[:,3]
#     yerrR = ft_stats[:,4] - ft_stats[:,3]
    xerrL = alert_stats[:,2]/sqrt(N)
    xerrR = alert_stats[:,2]/sqrt(N)
    yerrL = nmac_stats[:,2]/sqrt(N)
    yerrR = nmac_stats[:,2]/sqrt(N)
    
    
    PyPlot.errorbar(alert_stats[:, 1], nmac_stats[:, 1], 
    xerr = [xerrL xerrR ]',  marker="o", linestyle = "--", yerr = [yerrL yerrR]')
end
PyPlot.ylim([0, 1.5])
PyPlot.xlim([-0.05, 1.5])
PyPlot.grid("on")
PyPlot.xlabel("Alert Rate (commands/minute)")
PyPlot.ylabel("NMACs Per Flight Hour")
PyPlot.legend([string(n)*"-phase" for n in 1:length(data)], bbox_to_anchor=(0.74, 1.), loc = 9, ncol=2, numpoints=1)
#plt.savefig("./results/Flight_hours_VS_alert_rates.pgf", bbox_inches="tight")
```
