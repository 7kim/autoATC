```julia
colors = ["k","b", "y", "c", "r", "r", "k", "m","b","y"];
```

```julia
markers = ["", "o", "s", "D", "v", "^", "x"];
```

```julia
yold = -100;
for n in 2:length(data) #1 is the silent policy_prefix
    atcIdx = 1
    
    nBetas = size(data[n]["flightTimes"],1);
    N = size(data[n]["flightTimes"],3)

    nmac_stats = zeros(nBetas+1, 2) #mean and std
    alert_stats = zeros(nmac_stats);

    ismcts = data[n]["policy"] == "MCTS"
    
    for b in 1:(nBetas+1)
        
        nIdx = n
        betaIdx = b
        
        if b > nBetas #this is the silent run
            betaIdx = 1
            nIdx = 1
        end
        
        ft = data[nIdx]["flightTimes"][betaIdx,atcIdx,:][:]/3600
        nmacsPerHour  = data[nIdx]["nNMACcounts"][betaIdx,atcIdx,:][:]

        alertsPerHour = data[nIdx]["alertCounts"][betaIdx,atcIdx,:][:] ./ ft
        #plt.plot(alertsPerHour/60, nmacsPerHour, linestyle="", marker=".")        

        #nmacsPerHour = nmacsPerHour[!isinf(nmacsPerHour)] 
        #nmacsPerHour[isinf(nmacsPerHour)] = 20 #Just in case we had things without incidents...
        nmac_stats[b,:]  = [mean_and_std(nmacsPerHour)...]
        alert_stats[b,:] = [mean_and_std(alertsPerHour/60)...]
        
    end
    if nBetas == 1
        nmac_stats  = nmac_stats[1:(end-1), :]
        alert_stats = alert_stats[1:(end-1), :]
    end
    
    xerrL = 2*alert_stats[:,2]/sqrt(N)
    xerrR = 2*alert_stats[:,2]/sqrt(N)
    yerrL = 2*nmac_stats[:,2] /sqrt(N)
    yerrR = yerrL

    
    prefix = ""
    pfix = string(data[n]["nPhases"])*"-phase"
    
    xpos = alert_stats[1, 1]+.05
    ypos = nmac_stats[1, 1]
    if ismcts
        pfix *= "-mcts"
        #xpos -= .4
        ypos += .6
    elseif data[n]["policy"] == "GSMDP"
        pfix = "gsmdp"
    end
    #xpos = 1.25
    
    if(abs(ypos - yold) < 1)
        #ypos += .5
    end
    yold = ypos
    
    c = colors[n]
    m = markers[n]
    txt = prefix*pfix
    
    plt.text(xpos , ypos,  txt)#, color=c)

    ls = PyPlot.errorbar(alert_stats[:, 1], (nmac_stats[:, 1]), 
    xerr = [xerrL xerrR ]',  marker=m, linestyle = "--", yerr = [yerrL yerrR]')#, color=c)
    
    #plt.annotate(txt, [xpos, ypos], [1.3, ypos])


end
PyPlot.xlim([-0.05, 1.5])
PyPlot.grid("on")
PyPlot.xlabel("Alert Rate (commands/minute)")
PyPlot.ylabel("Mean Number of NMACs per 24 hours")

#PyPlot.legend([string(n)*"-phase" for n in 1:length(data)], bbox_to_anchor=(0.74, 1.), loc = 9, ncol=2, numpoints=1)
#plt.savefig("../CDC2015/CTMDP_MeanNMACs_VS_alert_rates.pgf", bbox_inches="tight")
```
