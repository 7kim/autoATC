```julia
linestyle = ["--","-"]
for (idx, policy_prefix) in enumerate(["greedyRandom_", ""])
    gca()[:set_color_cycle](["b","g", "r","c","m"])
    data = {load("./results/"*policy_prefix*"new_3DSimResults_n"*string(n)*".jld") for n in 1:5};
    # for n in 1:5
    #     push!(data, load("./results/steadystate_3DSimResults_n"*string(n)*".jld"))
    # end
    phaseNums = [1:5, 1:5]

    nBetas = size(data[1]["flightTimes"],1)
    N = size(data[1]["flightTimes"],3)


    nmac_stats = zeros(nBetas+1, 2) #mean and std
    alert_stats = similar(nmac_stats)


    minN = Inf
    maxN = -Inf
    for n in 1:length(data)
        for b in 1:(nBetas+1)
            atcIdx = 1
            betaIdx = b
            if b > nBetas
                atcIdx = 2 #Silent ATC
                betaIdx = 1
            end
            ft = data[n]["flightTimes"][betaIdx,atcIdx,:][:]/3600
            nmacsPerHour = ft ./ data[n]["nNMACcounts"][betaIdx,atcIdx,:][:]# ./ ft
            nmacsPerHour = nmacsPerHour[!isinf(nmacsPerHour)] #Just in case we had things without incidents...
            alertsPerHour = data[n]["alertCounts"][betaIdx,atcIdx,:][:] ./ ft
            nmac_stats[b,:]    = [mean_and_std(nmacsPerHour)...]
            alert_stats[b,:] = [mean_and_std(alertsPerHour/60)...]
        end
        #sort by alert means
        p = sortperm(alert_stats[:, 1]) #sorting using the mean
        alert_stats = alert_stats[p, :]
        nmac_stats = nmac_stats[p, :]

        minN = min(minN, minimum(nmac_stats[:,1]))
        maxN = max(maxN, maximum(nmac_stats[:,1]))

    #     xerrL = alert_stats[:,2] - alert_stats[:,3]
    #     xerrR = alert_stats[:,4] - alert_stats[:,3]

    #     yerrL = ft_stats[:,2] - ft_stats[:,3]
    #     yerrR = ft_stats[:,4] - ft_stats[:,3]
        xerrL = alert_stats[:,2]/sqrt(N)
        xerrR = alert_stats[:,2]/sqrt(N)
        yerrL = nmac_stats[:,2]/sqrt(N)
        yerrR = nmac_stats[:,2]/sqrt(N)

        PyPlot.errorbar(alert_stats[:, 1], nmac_stats[:, 1], 
        xerr = [xerrL xerrR ]',  marker="o", linestyle = linestyle[idx], yerr = [yerrL yerrR]')
        
        if(idx > 1)
            plt.text(alert_stats[end, 1]+.05, nmac_stats[end, 1]-.01,  string(phaseNums[n])*"-phase")
        end
    end
    #PyPlot.ylim([0, 1.5])
    PyPlot.xlim([-0.05, 1.8])
    PyPlot.grid("on")
    PyPlot.xlabel("Alert Rate (commands/minute)")
    PyPlot.ylabel("Mean Time to NMAC (hours)")

end


#PyPlot.legend([string(n)*"-phase" for n in 1:length(data)], bbox_to_anchor=(0.74, 1.), loc = 9, ncol=2, numpoints=1)
#plt.savefig("./results/CTMDP_MeanTimetoNMAC_VS_alert_rates.pgf", bbox_inches="tight")
```
