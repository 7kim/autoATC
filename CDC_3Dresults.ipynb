```julia
nBetas = size(data[1]["flightTimes"],1)
silentATCrun = size(data[1]["flightTimes"],2) == 2
N = size(data[1]["flightTimes"],3)

nmac_stats = zeros(nBetas+1, 2) #mean and std
alert_stats = zeros(nmac_stats);
```

```julia
for n in 1:length(data)
    atcIdx = 1
    for b in 1:(nBetas)
        betaIdx = b
        ft = data[n]["flightTimes"][betaIdx,atcIdx,:][:]/3600
        nmacsPerHour =  ft ./ data[n]["nNMACcounts"][betaIdx,atcIdx,:][:]# ./ ft

        alertsPerHour = data[n]["alertCounts"][betaIdx,atcIdx,:][:] ./ ft
        plt.plot(alertsPerHour/60, nmacsPerHour, linestyle="", marker=".")        

        #nmacsPerHour = nmacsPerHour[!isinf(nmacsPerHour)] #Just in case we had things without incidents...
        nmac_stats[b,:]    = [mean_and_std(nmacsPerHour)...]
        alert_stats[b,:] = [mean_and_std(alertsPerHour/60)...]
    end
    
    xerrL = alert_stats[:,2]/sqrt(N)*2
    xerrR = alert_stats[:,2]/sqrt(N)*2
    yerrL = nmac_stats[:,2]/sqrt(N)*2
    yerrR = nmac_stats[:,2]/sqrt(N)*2

    PyPlot.errorbar(alert_stats[:, 1], nmac_stats[:, 1], 
    xerr = [xerrL xerrR ]',  marker="o", linestyle = "--", yerr = [yerrL yerrR]')
end
grid("on")
```
