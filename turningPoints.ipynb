```julia
function tr2mat(tr)
    n = length(tr)
    M = Array(Float64, 3, n)
    for (i, p) in enumerate(tr)
        M[1, i] = p.n
        M[2, i] = p.e
        M[3, i] = p.psi
    end
    return M
    
end
using MAT
```

```julia
function movingAverage(y, n)
    assert(n%2  == 0)
    k = int(n/2)

    mavg = ones(Float64, n) / n
    y_filt = [fill(y[1],k), y, fill(y[end],k)]
    y_filt = conv(y_filt, mavg)
    y_filt = y_filt[n:(end-n)]
    
    assert(size(y) == size(y_filt))
    
    return y_filt
end
```

```julia
function splits(arr, fun)
    idx = find(fun, arr)
    didx = [0, find(i -> i > 2, diff(idx))]
    push!(didx,length(idx))
    
    spl = Vector{Int64}[]
    for k in 1:(length(didx)-1)
        push!(spl, idx[(didx[k]+1):didx[k+1]])
    end
    return spl
end
```

```julia
headings = Array(Vector{Float64},length(trajectories));
```

```julia
filtHeadings = similar(headings)
diffHeadings = similar(headings);
diffHeadings_smooth = similar(headings);
tp_n = similar(headings);
tp_e = similar(headings);

all_n = similar(headings);
all_e = similar(headings);
```

```julia
for (tidx, t) in enumerate(trajectories)
    n = Float64[mes.n  for mes in t]; all_n[tidx] = n
    e = Float64[mes.e  for mes in t]; all_e[tidx] = e 
    dn = diff(n); push!(dn, dn[end]) #push! to keep same size 
    de = diff(e); push!(de, de[end])
    
    
    headings[tidx] =  atan2(de, dn)
    
    
    ca = movingAverage(cos(headings[tidx]),8)
    sa = movingAverage(sin(headings[tidx]),8)
    filtHeadings[tidx] = atan2(sa, ca)
    diffHeadings[tidx] = abs(map(unwrap,diff(filtHeadings[tidx])))
    diffHeadings_smooth[tidx] = movingAverage(diffHeadings[tidx],4)
    
    #find region where heading derivative is > thresh
    spl = splits(diffHeadings_smooth[tidx], x -> x >= 0.2)
    
    n_tp = Float64[]
    e_tp = Float64[]
    for idx in spl
        if length(idx) > 0
            idx0 = floor(int(middle(idx)))
            push!(n_tp,n[idx0])
            push!(e_tp,e[idx0])
        end
    end
    tp_n[tidx] = n_tp
    tp_e[tidx] = e_tp

end
#This is all of the data
XBig = [ vcat(all_e...)', vcat(all_n...)', cos(vcat(filtHeadings...)'), sin(vcat(filtHeadings...)')]';

#This is only turning points
X = [vcat(tp_n...)', vcat(tp_e...)']
```

```julia
χ_2 = Chisq(2);
function confidenceInterval(g, Xs)
    l_x = g[:score](Xs)
    μ = g[:means_]
    l_μ = g[:score](μ)[1]
    #Only using χ_2 for now
    @assert length(μ) == 2
    
    #compute the Q value = (x-μ)Σ_inv(x-μ)
    xmSmx = 2*(l_μ - l_x)
    
    #figure out cdf value
    #This is the probability that a random sample
    #will lie inside of an ellipse right at the edge of this point Xs
    c = cdf(χ_2, xmSmx)
    
    #1-c -> 1 as Xs -> μ and -> 0 as Xs is far from it
    return 1-c
end
```

```julia
trajLabels = Array(Vector{typeof([0,0])}, length(trajectories))

for (tidx, t) in enumerate(trajectories)
    nList = Float64[mes.n  for mes in t];
    eList = Float64[mes.e  for mes in t];
    Xi = [vcat(eList...)', vcat(nList...)']'
    
    tLength = length(nList)
    
    scores = Array(Float64, tLength, length(label_unique))
    for (lidx, l) in enumerate(label_unique)
        scores[:, lidx] = confidenceInterval(gDict[l],Xi)
    end
    
    maxscores, maxidx = findmax(scores, 2)
    #change idx into label indx
    maxidx = Int64[ind2sub(size(scores),i)[2] for i in maxidx]
    
    #use 0 based labels
    maxidx -= 1;

    #Remove anything below a 0.1 threshold
    low_idx = find(x -> x < 0.1, maxscores)
    maxidx[low_idx] = -1;
    
    
    trajLabels[tidx] = [[m,m] for m in maxidx]
    
    #Populate the from field
    l = trajLabels[tidx][1][1];
    for j in 2:tLength
        lnew = trajLabels[tidx][j][1]
        if lnew != l && lnew != -1
            l = lnew
        end
        trajLabels[tidx][j][1] = l
    end
    #Populate the to field
    l = trajLabels[tidx][end][2];
    for j in (tLength-1):-1:1
        lnew = trajLabels[tidx][j][2]
        if lnew != l && lnew != -1
            l = lnew
        end
        trajLabels[tidx][j][2] = l
    end
#     plot(trajLabels[tidx])
    
end
# grid("on")
```

```julia
counts = zeros(length(label_unique), length(label_unique))
for traj in trajLabels
    
    state = :LEG
    for from_to in traj
        from = from_to[1]+1;
        to = from_to[2]+1;
        if(from == 0 || to == 0)
            continue 
        end
        
        
        if(state == :LEG)
            if(from == to)
                state = :NODE
            end
        elseif(state == :NODE)
            if(from != to)
                #we just transitioned
                state = :LEG
                counts[from, to] += 1
            end
        end        
    end
    
end
```

```julia
sums_inv = 1./sum(counts,2)
sums_inv[isinf(sums_inv)]=1.
D = diagm(vec(sums_inv))
probs = D * counts
```
