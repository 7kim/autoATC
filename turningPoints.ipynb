```julia
function tr2mat(tr)
    n = length(tr)
    M = Array(Float64, 3, n)
    for (i, p) in enumerate(tr)
        M[1, i] = p.n
        M[2, i] = p.e
        M[3, i] = p.psi
    end
    return M
    
end
using MAT
```

```julia
m = [tr2mat(tr) for tr in trajectories];
file = matopen("trajectories_sim.mat", "w")
write(file, "trajectories", m)
close(file)
```

```julia
function movingAverage(y, n)
    assert(n%2  == 0)
    k = int(n/2)

    mavg = ones(Float64, n) / n
    y_filt = [fill(y[1],k), y, fill(y[end],k)]
    y_filt = conv(y_filt, mavg)
    y_filt = y_filt[n:(end-n)]
    
    assert(size(y) == size(y_filt))
    
    return y_filt
end
```

```julia
function splits(arr, fun)
    idx = find(fun, arr)
    didx = [0, find(i -> i > 2, diff(idx))]
    push!(didx,length(idx))
    
    spl = Vector{Int64}[]
    for k in 1:(length(didx)-1)
        push!(spl, idx[(didx[k]+1):didx[k+1]])
    end
    return spl
end
```

```julia
headings = Array(Vector{Float64},length(trajectories));
```

```julia
filtHeadings = similar(headings)
diffHeadings = similar(headings);
diffHeadings_smooth = similar(headings);
tp_n = similar(headings);
tp_e = similar(headings);

all_n = similar(headings);
all_e = similar(headings);
```

```julia
tmp = diff(all_n[1]); push!(tmp, tmp[end])
```

```julia
for (tidx, t) in enumerate(trajectories)
    n = Float64[mes.n  for mes in t]; all_n[tidx] = n
    e = Float64[mes.e  for mes in t]; all_e[tidx] = e 
    dn = diff(n); push!(dn, dn[end]) #push! to keep same size 
    de = diff(e); push!(de, de[end])
    
    
    headings[tidx] =  atan2(de, dn)
    
    
    ca = movingAverage(cos(headings[tidx]),8)
    sa = movingAverage(sin(headings[tidx]),8)
    filtHeadings[tidx] = atan2(sa, ca)
    diffHeadings[tidx] = abs(map(unwrap,diff(filtHeadings[tidx])))
    diffHeadings_smooth[tidx] = movingAverage(diffHeadings[tidx],4)
    
    #find region where heading derivative is > thresh
    spl = splits(diffHeadings_smooth[tidx], x -> x >= 0.2)
    
    n_tp = Float64[]
    e_tp = Float64[]
    for idx in spl
        if length(idx) > 0
            idx0 = floor(int(middle(idx)))
            push!(n_tp,n[idx0])
            push!(e_tp,e[idx0])
        end
    end
    tp_n[tidx] = n_tp
    tp_e[tidx] = e_tp

end
#This is all of the data
XBig = [ vcat(all_e...)', vcat(all_n...)', cos(vcat(filtHeadings...)'), sin(vcat(filtHeadings...)')]';

#This is only turning points
X = [vcat(tp_n...)', vcat(tp_e...)']
```

```julia
Xdict_keys = keys(Xdict)
trajLabels = Array(Vector{typeof([0,0])}, length(trajectories))
for (tidx, t) in enumerate(trajectories)
    nList = Float64[mes.n  for mes in t];
    eList = Float64[mes.e  for mes in t];
   
    tLength = length(nList)
    
    #Initialze the labels to (-1,-1)
    trajLabels[tidx] = [[-1,-1] for i in 1:tLength]
    #flag the keys 
    for (idx, (n, e)) in enumerate(zip(nList, eList))
        k = Float64[n, e];
        if k in Xdict_keys
            l = Xdict[k]
            trajLabels[tidx][idx] = [l,l];
            
            #Populate the to field
            j = idx-1
            while j > 1
                if trajLabels[tidx][j][2] != -1
                    break;
                end
                trajLabels[tidx][j][2] = l
                j -= 1; 
            end
        end
    end

    #Populate the from field
    l = trajLabels[tidx][1][1];
    for j in 2:tLength
        lnew = trajLabels[tidx][j][1]
        if lnew != l && lnew != -1
            l = lnew
        end
        trajLabels[tidx][j][1] = l
    end
end
```
