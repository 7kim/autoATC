```julia
function tr2mat(tr)
    n = length(tr)
    M = Array(Float64, 3, n)
    for (i, p) in enumerate(tr)
        M[1, i] = p.n
        M[2, i] = p.e
        M[3, i] = p.psi
    end
    return M
    
end
using MAT
```

```julia
m = [tr2mat(tr) for tr in trajectories];
file = matopen("trajectories_sim.mat", "w")
write(file, "trajectories", m)
close(file)
```

```julia
function movingAverage(y, n)
    assert(n%2  == 0)
    k = int(n/2)

    mavg = ones(Float64, n) / n
    y_filt = [fill(y[1],k), y, fill(y[end],k)]
    y_filt = conv(y_filt, mavg)
    y_filt = y_filt[n:(end-n)]
    
    return y_filt
end
```

```julia
function splits(arr, fun)
    idx = find(fun, arr)
    didx = [0, find(i -> i > 2, diff(idx))]
    push!(didx,length(idx))
    
    spl = Vector{Int64}[]
    for k in 1:(length(didx)-1)
        push!(spl, idx[(didx[k]+1):didx[k+1]])
    end
    return spl
end
```

```julia
headings = Array(Vector{Float64},length(trajectories));
```

```julia
filtHeadings = similar(headings)
diffHeadings = similar(headings);
diffHeadings_smooth = similar(headings);
tp_n = similar(headings);
tp_e = similar(headings);
```

```julia
for (tidx, t) in enumerate(trajectories)
    n = Float64[mes.n  for mes in t];
    e = Float64[mes.e  for mes in t];
    dn = diff(n);
    de = diff(e);
    headings[tidx] =  atan2(de, dn)
    
    
    ca = movingAverage(cos(headings[tidx]),8)
    sa = movingAverage(sin(headings[tidx]),8)
    filtHeadings[tidx] = atan2(sa, ca)
    diffHeadings[tidx] = abs(map(unwrap,diff(filtHeadings[tidx])))
    diffHeadings_smooth[tidx] = movingAverage(diffHeadings[tidx],4)
    
    #find region where heading derivative is > thresh
    spl = splits(diffHeadings_smooth[tidx], x -> x >= 0.2)
    
    n_tp = Float64[]
    e_tp = Float64[]
    for idx in spl
        if length(idx) > 0
            idx0 = floor(int(middle(idx)))
            push!(n_tp,n[idx0])
            push!(e_tp,e[idx0])
        end
    end
    tp_n[tidx] = n_tp
    tp_e[tidx] = e_tp
end
X = [vcat(tp_n...)', vcat(tp_e...)']
```

```julia
file = matopen("turning_points_sim.mat", "w")
write(file, "X", X)
close(file)
```

```julia
for i = 200
    plot(headings[i],marker=".",markersize=1, linestyle="", color="blue")
    plot(filtHeadings[i],color="blue")
    
    spl = splits(diffHeadings_smooth[i], x -> x >= .15)
    plot(diffHeadings_smooth[i]*10, color="red", linestyle="--")
    for idx in spl
        plot(idx,diffHeadings_smooth[i][idx]*10, color="red")
        
        idx0 = floor(int(middle(idx)))
        plot(idx0, filtHeadings[i][idx0],marker="x",color="red")
    end
end
grid("on")
```
